```python
import pandas as pd
import jieba
import gensim
from gensim import corpora
import pyLDAvis.gensim

# 加载停用词表
stopwords = []
with open('D:/pyhton_practice/Teddy cup/stopwords.txt', 'r', encoding='utf-8') as f:
    for line in f.readlines():
        stopwords.append(line.strip())

# 对文本进行分词和去停用词
def tokenizer(text):
    return [word for word in jieba.lcut(text) if word not in stopwords]

# 对职位需求列进行分词和去停用词
df['职位需求'] = df['职位需求'].apply(tokenizer)

# 构建词典和语料库
dictionary = corpora.Dictionary(df['职位需求'])
corpus = [dictionary.doc2bow(text) for text in df['职位需求']]

# 使用 LDA 模型进行主题提取
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                            id2word=dictionary,
                                            num_topics=10,  # 设定主题的数量
                                            passes=10)  # 设定迭代次数

# 输出每个主题的关键词
topics = lda_model.show_topics(num_topics=10, num_words=10, formatted=False)
for i, topic in enumerate(topics):
    print('主题 {} : {}'.format(i+1, [word[0] for word in topic[1]]))
    
# 可视化
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```

```python
import pandas as pd
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer

# 读取数据并去除符号
df['福利补贴'] = df['福利补贴'].str.replace('[^\u4e00-\u9fa5]+', '')
```

```python
# 加载停用词表
stopwords = []
with open('D:/pyhton_practice/Teddy cup/stopwords.txt', 'r', encoding='utf-8') as f:
    for line in f.readlines():
        stopwords.append(line.strip())

# 添加自定义词汇
jieba.add_word('五险一金')

# 对文本进行分词和去停用词
def tokenizer(text):
    return [word for word in jieba.cut(text) if word not in stopwords]

df['福利补贴'] = df['福利补贴'].apply(tokenizer)

# 计算tfidf并获取前十个词的词频和tfidf值
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['福利补贴'].apply(lambda x: ' '.join(x)))
feature_names = tfidf.get_feature_names()
tfidf_scores = tfidf_matrix.sum(axis=0).A1
results = dict(zip(feature_names, tfidf_scores))

# 统计词频并获取前十个词的词频和tfidf值
word_freq = {}
for text in df['福利补贴']:
    for word in text:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
top_10_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:15]

# 输出结果
for word, freq in top_10_words:
    print('词语：', word)
    print('词频：', freq)
    print('tfidf值：', results[word])
    print('--------------')
```
