```python
data_id = pd.read_csv('D:/pyhton_practice/TeddyCup_C/data/result1-2-2_test.csv')
data_id = data_id.loc[:, ['招聘信息ID', '职位关键词']]

# 将 data_id 中的招聘信息ID列转换为 int 类型
data_id['招聘信息ID'] = data_id['招聘信息ID'].astype('int64')

# 修改 data_id 的列名，将“招聘信息ID”修改为“招聘信息 ID”
data_id = data_id.rename(columns={'招聘信息ID': '招聘信息ID'})

# 将两个 DataFrame 按照招聘信息 ID 进行合并
merged_data = pd.merge(data_total, data_id, on='招聘信息ID')

# 将 data_id 中的职位关键词列转移至 merged_data 中，并重新命名为“职位关键词”
merged_data['职位关键词'] = merged_data['职位关键词'].combine_first(merged_data['职位关键词'])

# 删除原始的“职位关键词_x”和“职位关键词_y”列
# merged_data.drop(['职位关键词_x', '职位关键词_y'], axis=1, inplace=True)

# 保存结果到新的 csv 文件中
merged_data
```

```python
# 使用 range() 函数生成序号并赋值给 '序号' 列
merged_data.drop(columns=['序号'], inplace=True)
merged_data.insert(0, '序号', range(1, len(merged_data)+1))
# 交换"企业名称"和"招聘岗位"的位置
merged_data = merged_data.reindex(columns=['序号', '招聘信息ID', '企业名称', '招聘岗位', 
                                           '职位发布', '投递截止', '岗位性质','技能关键词', '职位关键词', '职位需求', '工作经验要求',
                                          '最低学历要求', '最少工资', '最大工资', '工资支付方式', '福利补贴',
                                          '职称', '工作地点', '招聘人数', '详细地址', '企业简称', '领域方向', '企业性质',
                                          '人员规模', '已投递简历数'])
```
