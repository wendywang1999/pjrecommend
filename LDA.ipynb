```python
import pandas as pd
import jieba
import gensim
from gensim import corpora
import pyLDAvis.gensim

# 加载停用词表
stopwords = []
with open('D:/pyhton_practice/Teddy cup/stopwords.txt', 'r', encoding='utf-8') as f:
    for line in f.readlines():
        stopwords.append(line.strip())

# 对文本进行分词和去停用词
def tokenizer(text):
    return [word for word in jieba.lcut(text) if word not in stopwords]

# 对职位需求列进行分词和去停用词
df['职位需求'] = df['职位需求'].apply(tokenizer)

# 构建词典和语料库
dictionary = corpora.Dictionary(df['职位需求'])
corpus = [dictionary.doc2bow(text) for text in df['职位需求']]

# 使用 LDA 模型进行主题提取
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                            id2word=dictionary,
                                            num_topics=10,  # 设定主题的数量
                                            passes=10)  # 设定迭代次数

# 输出每个主题的关键词
topics = lda_model.show_topics(num_topics=10, num_words=10, formatted=False)
for i, topic in enumerate(topics):
    print('主题 {} : {}'.format(i+1, [word[0] for word in topic[1]]))
    
# 可视化
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```

```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Scatter, Line

# 读取数据
df = pd.read_csv('D:/pyhton_practice/TeddyCup_C/data/result1-1_test.csv')

# 统计年龄人数
age_counts = df['年龄'].value_counts().sort_index()

# 绘制散点图
scatter = Scatter()
scatter.add_xaxis(age_counts.index.tolist())
scatter.add_yaxis("", age_counts.tolist())
scatter.set_global_opts(title_opts=opts.TitleOpts(title="年龄分布"),
                        xaxis_opts=opts.AxisOpts(name="年龄"),
                        yaxis_opts=opts.AxisOpts(name="人数"))

# 线性拟合
x = age_counts.index.tolist()
y = age_counts.tolist()
a, b = np.polyfit(x, y, 1)
line_y = [a * xi + b for xi in x]

# 添加拟合后的曲线
line = Line()
line.add_xaxis(x)
line.add_yaxis("", line_y, is_smooth=True)

# 将散点图和曲线组合在一起
scatter.overlap(line)

# 展示图表
scatter.render_notebook()
```
